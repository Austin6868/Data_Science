Importing all the libraries...

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="darkgrid")
```

import the file

```python
df = pd.read_csv('fortune500.csv')
```

Changing the column names to more workable ones

```python
df.columns = ['year', 'rank', 'company', 'revenue', 'profit']
```

Check if we have the right length

```python
len(df)
```

Check the datatypes, here we have an unoptimized datatype 'object' for profit which is supposed to be a float

```python
df.dtypes
```

Therefore we would use regex to detect and show which items is not parsed as a number '^' here means 'not' in boolean operations

```python
non_numberic_profits = df.profit.str.contains('[^0-9.-]')
df.loc[non_numberic_profits].head()
```

As we see, the only kind not correct possibility is 'N.A.'

```python
set(df.profit[non_numberic_profits])
```

And we have 369 'N.A.'s that we need to take care of

```python
len(df.profit[non_numberic_profits])
```

Try and see how are the invalid data distributed and most of them are around 1990 area

```python
bin_sizes, _, _ = plt.hist(df.year[non_numberic_profits], bins=range(1955, 2006))
```

Here we remove all the rows that have invalid profits

```python
df = df.loc[~non_numberic_profits]
df.profit = df.profit.apply(pd.to_numeric)
```

Check if we got it right

```python
len(df)
```

Again, if we got it right with types...

```python
df.dtypes
```

Getting ready to plot a line graph with the data we have, we are doing both profit and revenue respectively

```python
group_by_year = df.loc[:, ['year', 'revenue', 'profit']].groupby('year')
avgs = group_by_year.mean()
x = avgs.index
y1 = avgs.profit
def plot(x, y, ax, title, y_label):
    ax.set_title(title)
    ax.set_ylabel(y_label)
    ax.plot(x, y)
    ax.margins(x=0, y=0)
```

Making the graph itself...

```python
fig, ax = plt.subplots()
plot(x, y1, ax, 'Increase in mean Fortune 500 company profits from 1955 to 2005', 'Profit (millions)')
```

Doing the same for revenue...

```python
fig, ax = plt.subplots()
y2 = avgs.revenue
plot(x, y2, ax, 'Increase in mean Fortune 500 company profits from 1955 to 2005', 'Profit (millions)')
```

Here we are doing something fun, plotting with standard deviation using the fill_between() functionality

```python
def plot_with_std(x, y, stds, ax, title, y_label):
    ax.fill_between(x, y - stds, y + stds, alpha=0.2)
    plot(x, y, ax, title, y_label)
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2)
title = 'Increase in mean and std Fortune 500 company %s from 1955 to 2005'
# stds1 is for profit
stds1 = group_by_year.std().profit.values
# stds2 is for revenue
stds2 = group_by_year.std().revenue.values
plot_with_std(x, y1.values, stds1, ax1, title % 'profits', 'Profit (millions)')
plot_with_std(x, y2.values, stds2, ax2, title % 'revenues', 'Revenue (millions)')
fig.set_size_inches(20, 4)
fig.tight_layout()
```
